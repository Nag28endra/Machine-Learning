```python
import seaborn as sns
# importing titanic dataset
titanic = sns.load_dataset('titanic')

titanic.head(5)
```

The dataset has 15 columns with 891 non-null values.

Great! we got some statistical data.

It seems like our dataset don't have any missing values. Why don't we create some missing values.

let's do that from the below code.

```python
import numpy as np
# generate some random seed
np.random.seed(42)

#create a mask
mask = np.random.rand(*titanic.shape)<0.2

# apply the mask to the dataset to create missing values.
titanic_missing = titanic.mask(mask)
titanic_missing.info()
```

Barvooo!! You can see that there are so many missing values in our dataset right now 😅.

Let's understand our code now.

A **random mask** is a boolean array, or matrix, in which each element has a probability of being either *True* or False when random missing values are introduced. To determine which values in a dataset should have missing values (usually NaN) substituted, utilize this boolean array.


A random array with the same shape as the Titanic dataset is created by running **`np.random.rand(*titanic.shape)`**. Values in the array range from 0 to 1. A threshold of **< 0.2** produces a boolean mask whose elements are True if the corresponding random value is less than 0.2. This threshold can be adjusted based on the desired proportion of missing values.

Wherever the corresponding element in the mask is True, the **`mask()`** function substitutes NaN for the values in the original dataset (titanic).




### <a id="handling-missing-data" name="handling-missing-data">3.3.1 Handling missing data</a>

Let's display the count of missing values in each column of our dataset.

```python
# displaying the missing values count in the each row

titanic_missing.isnull().sum()
```

**`< dataset>.isnull().sum()`** method gives you the count of missing values in each column.

We can handle missing values either by removing the rows with null values or by imputing some other values.

We'll see both of them.

Let's drop a few rows of specific columns. Check the code given below;

```python
# dropping rows with values  'Null' in the 'who' column

titanic_missing= titanic_missing.dropna(subset=['who'])
titanic_missing.isnull().sum()
```

From the above code we have dropped all the rows having null values in the 'who' column. For this, we have used **`dropna()`** method. We have passed the **`subset`** argument to the method to drop only certain column related data.

Rather than dropping those null values, we can fill those values using some mean or median or mode values. This is what we call '**`imputing`**'.

This can done using **`fillna()`** method.

```python
titanic_missing.info()
```

```python
# filling 'survived' and 'age' column null values with mean values.

# specifying the columns to be filled.
specific_columns = ['survived','age']
# filling the missing values in the 'survived' and 'age' column.
titanic_missing[specific_columns] = titanic_missing[specific_columns].fillna(titanic_missing[specific_columns].mean())

# counting the missing values in each column.
titanic_missing.isnull().sum()
```

For the above code, we have imputed mean values to only two column's missing values. For that, we have only selected those columns and made changes to those columns instead of making changes for the entire dataset.

You can check there are no null values in the 'survived' and 'age' columns.

Let's do it for the remaining columns but this time using 'median' values.

```python
# replacing null values with 'median' values.
specific_columns = ['pclass','sibsp','parch','fare']

# replacing only specific values using median() method.
titanic_missing[specific_columns] = titanic_missing[specific_columns].fillna(titanic_missing[specific_columns].median())

# checking the missing values.
titanic_missing.isnull().sum()
```

Great!! we have successfully replaced the null values with 'mean' and 'median' values for those specific columns.

But you see, there are a few columns still having some null values. But we cannot replace them with mean or median or mode.

```python
titanic_missing[['sex','embarked','class','adult_male','deck','embark_town','alive','alone']].dtypes
```

The above code shows you the data types of other column's missing or null values. They are mostly categorical and object(string values.)

We need to handle the categorical data. We'll see this in the next section.

### <a id="categorical-outliers" name="categorical-outliers">3.3.2 Handling categorical data and outliers</a>

In this section, we'll learn how to handle the remaining missing values in our dataset which are mostly categorical and objects.

Handling missing values for categorical data contains many strategies. But for now, we'll use very simple ones. Later in other chapters, we'll explore different methods and strategies.

```python
# choosing the columns
category_labels = ['sex','embarked','class','adult_male','deck','embark_town','alive','alone']

#replacing the columns with mode value.
titanic_missing[category_labels] =titanic_missing[category_labels].fillna(titanic_missing[category_labels].mode().iloc[0])

# copying the clean data with no missing values.
titanic_clean = titanic_missing

#checking the null values.
titanic_clean.isnull().sum()
```

Good to go, we have cleared all the null values or missing values that we have created.

We have replaced the category and object-based columns with the 'mode' strategy. The purpose of **`iloc[0]`** is to extract the first (most frequent) value from the mode of the 'categorical_column'.

When it comes to outliers, data points that significantly deviate from the bulk of the data in a dataset are called **outliers**. These data points may cause the overall statistical analysis to be distorted because they are abnormally high or low about the other values. Errors in data collection, measurement errors, or true extreme values in the underlying distribution are some of the causes of outliers.

We can figure out outliers by visualizations. We'll go through it when we go to other chapters.

### <a id="aggregation" name="aggregation">3.3.4 Aggregation functions using Pandas</a>

Aggregation functions in Pandas are used to operate on data, usually transforming or summarizing data across rows or columns. These features support the analysis and comprehension of the data's properties.



Some common aggregation functions are;

* **mean()**: Calculate the mean (average) of each group.
* **sum()**: Compute the sum of values for each group.
* **min(), max()**: Find the minimum or maximum value in each group.
* **count()**: Count the number of non-null values in each group.
* **size()**: Count the total number of values in each group (including null values).

```python
# Example: Calculate mean, sum, and count for a specific column
raw_data=titanic_clean.groupby('class')['age'].agg(['mean', 'sum', 'count'])

raw_data
```

This example computes the mean, sum, and count of the '**`age`**' column for each passenger class using the '**`class`**' column as the grouping criterion. Each passenger class will have one row in the resulting DataFrame, and the three columns ('mean, sum, and count') will show the corresponding aggregation results.



```python
result = titanic_clean.groupby('class').agg({'age': 'mean', 'fare': 'sum'})

result
```

We can also compute for multiple columns.

The grouping criterion is the 'class' column. According to the aggregation dictionary, the mean function should be used to aggregate the 'age' column and the sum function should be used to aggregate the 'fare' column. One row for each passenger class and two columns ('age' and 'fare') displaying the associated aggregation results will make up the final DataFrame.


Let's try out other aggreation functions.

```python
# counting the age related data in each passenger class
count_class = titanic_clean.groupby('class')['age'].agg(['count'])

count_class
```

### <a id="basic-plot" name="basic-plot">3.4.1 Basic plots.</a>

There are a lot of plots that we are going to discuss in this section.

Let's start with the line plot.

Let's plot some data.

* **Line plot**

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# giving labels
plt.xlabel('X-axis')
plt.ylabel('Y-axis',rotation=0)
plt.title('Line plot')
# Create a line plot
plt.plot(x, y)

# giving plot size
plt.figure(figsize=(3,4))

plt.show()
```

The code doesn't take much time to understand. We use the **`plot()`** method to plot the data. And we have **`xlabel()`** and **`ylabel()`** to set the labels for the axis. Using figure(figsize) we can set the figure size.

Using **`plt.show()`** we can display the plot.

* **Scatter plot**

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y, color='red', marker='o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

There is no need for an explanation of the code. The code itself is readable.

* **Bar graph.**

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 1, 5]

plt.bar(categories, values, color='blue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

The categories labeled 'A', 'B', 'C', and 'D' are represented by bars with heights of 3, 7, 1, and 5, respectively, in this bar plot created by Matplotlib. The bars are visually distinguished by their blue color. The plot is called "**Bar Plo**t," the y-axis is labeled "Values," and the x-axis is labeled "Categories." This simple visualization facilitates the interpretation of data distributions or trends by enabling rapid comparison of values associated with various categories.

Instead of using **`plt.plot()`** we use **`plt.bar()`** to plot the bar graph.

* **Histogram**

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]

plt.hist(data, bins=5, color='skyblue', edgecolor='red')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

We use the **`plt.hist()`** method to plot the histogram. We have the following arguments to pass to this method;

* data
* bins (number of groups.)
* color
* edge color

* **Pie chart**

```python
labels = ['Javascript', 'python', 'Java']
sizes = [25, 40, 35]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['orange', 'lightblue', 'lightgreen'])
plt.title('Pie Chart')
plt.show()
```

This code generates a pie chart that shows the distribution of programming language preferences using Matplotlib. The terms 'Javascript,' 'Python,' and 'Java' are associated with respective pie chart slices that account for 25%, 40%, and 35% of the total. Python is light blue, Java is light green, and Javascript is orange in this visually appealing pie chart. The first slice's starting angle is adjusted with the `startangle=90' parameter, while the 'autopct` parameter shows the percentage on each slice. The resulting pie chart makes it easier to understand which programming languages are more popular than others by providing a clear visual representation of their proportionate usage.

### <a id="Plot-customization" name="Plot-customization">3.4.2  Plot customization ( labels,titles,legends,colours)</a>

 Let's do some customizations to our plot.

 Consider the line plot for this section. We'll do various customizations to our plot.

```python
import numpy as np

# Create an array of 100 evenly spaced values from 0 to 10
x = np.linspace(0, 10, 100)
# Calculate the sine values for the x values
y = np.sin(x)
```

**`linspace()`** is an in-built function in Python's NumPy library. It is used to create an evenly spaced sequence in a specified interval.

```python
#plot the graph
plt.plot(x, y)
plt.show()
```

* adding labels to the plot

```python
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Line Plot')
plt.show()
```

* changing line color and style

```python
plt.plot(x, y, color='red', linestyle='--')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Custom Color and Style')
plt.show()
```

I can tell the code is readable and there is no need for an explanation for the given code.

* Adding grid lines

```python
plt.plot(x, y)
plt.grid(True)  # Add gridlines
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Gridlines')
plt.show()
```

* Changing line width

```python
plt.plot(x, y, linewidth=2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Custom Line Width')
plt.show()
```

* Adding annotations.

```python
plt.plot(x, y)
plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2, 1.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=12)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Annotation')
plt.show()
```

**`plt.annotate('Peak', xy=(np.pi/2, 1), xytext=(np.pi/2, 1.5),
arrowprops=dict(facecolor='black', arrowstyle='->'),fontsize=12)`**

This line of code uses the annotate function in Matplotlib to add an annotation to the line plot. Let's break down each argument:

'**`Peak`**': This is the text of the annotation, which will be displayed on the plot.

**`xy=(np.pi/2, 1)`**: Specifies the point on the plot where the arrow will point to. In this case, it's located at x-coordinate np.pi/2 (which is π/2, representing 90 degrees) and y-coordinate 1.

**`xytext=(np.pi/2, 1.5)`**: Specifies the position of the text label ('Peak'). It's similar to xy, but the text will be placed at a slightly higher y-coordinate of 1.5.

**`arrowprops=dict(facecolor='black', arrowstyle='->')`**: Configures the appearance of the arrow that points to the annotated point. It specifies that the arrow should be black (facecolor='black') and have an arrowhead (arrowstyle='->').

fontsize=12: Sets the font size of the text to be displayed.
