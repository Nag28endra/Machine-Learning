### <a id="DTA" name="DTA">9.1.1 Decision Tree algorithm</a>

<a href = "#top">⇑ table of contents</a>

- Decision trees are used for both regression and classification kinds of problems.

- They use tree-like structures and a combination of them to solve a problem.

- A decision tree is a non-linear structure that consists of root nodes, branches, and leaf nodes. Each internal node denotes a test on an attribute the branches denote the test outcomes and the leaf node contains the class labels.



image.png

And we'll make a few assumptions while implementing decision tree algorithms. They are;

- Initially, the training set of the dataset is considered as the root of the node.

- The features in the dataset need to be categorical. If not they are discretized before building the model.

- Rows are distributed recursively based on the attribute values.

- And at last, which order of attributes are placed in decision tree are done using some statistical approach.

### <a id="cart" name="cart">9.1.2 CART and its terminology</a>
<a href = "#top">⇑ table of contents</a>

The decision tree algorithm is also called as ***CART*** which means ***Categorical and Regression Trees***.

CART algorithm provides the foundation for other algorithms like **random** **forest**, **bagged decision trees**, and **boosted decision trees**.

Since decision trees are used for both classification and regression it is termed as CART. In this chapter, we are going to deal with a classification kind of problem.

Let's look at some of the terminologies involved in decision trees.

- **Root node**
> It represents the entire training set or the population of the dataset. It can further divide into two or more homogenous sets.

- **Splitting**
> It is the process of dividing the node into two or more subsets/ sub-nodes

- **Decision node**
> The internal node of the decision tree that holds some conditions is called the decision node.

- **Terminal or leaf node**
> These nodes that do not split further are the leaf or terminal nodes.

- **Pruning**
> The process of removing the sub-nodes of the tree is called pruning. It is the opposite process of splitting.



### <a id="intuition" name="intuition">9.1.3 Decision tree intuition</a>

<a href = "#top">⇑ table of contents</a>

One of the most popular and extensively applied supervised machine learning algorithms, the Decision-Tree algorithm can be used to both regression and classification problems. The Decision-Tree algorithm's intuition is extremely easy to comprehend.

The following is the intuition of the Decision Tree algorithm:


- By using the Decision-Tree algorithm, a node is formed for every attribute in the dataset. The root node contains the most crucial property.

- In order to evaluate the task at hand, we begin at the root node and follow the corresponding node that satisfies our condition or decision as we move down the tree.

- This process keeps going until it reaches a leaf node. It includes the Decision Tree's prediction or result.


## <a id="concepts" name="concepts">9.2 Decision Tree concepts</a>

<a href = "#top">⇑ table of contents</a>

Let's see some of the important concepts that the decision tree follows. We have already come to know about its intuition that it considers every attribute of the data set as a node and selects one attribute as a root node based on some statistical methods. We'll learn them in this section.

We'll select those attributes based on some attribute selection measures. Let's see those.

The primary challenge of the decision tree algorithm is to find the best attribute that can fit as a root node at each level of the decision tree. We'll choose the following methods for the attribute selection.

- Information gain
- Gini index

### <a id="information-gain" name="information-gain">9.2.1 Information gain </a>

<a href = "#top">⇑ table of contents</a>

When opting for the *information gain* criterion then we assume the attributes as categorical variables. To know more about the information gain we need to know about another concept called ***entropy***.

#### **Entropy**

- Entropy measures the randomness or the impurity in the dataset. It measures the uncertainty of a random variable X(feature).

- So if the entropy is less then the information gain will be more.

- Information gain is the difference between entropy before the split and average entropy after the split.

Entropy is formally given as;

image.png

Where 𝙥ᵢ is the probability associated with ith class and c is the number of classes.

And there is an algorithm called ***Iterative Dichotomiser(ID3)*** that uses entropy to calculate the information gain. It calculates the entropy of each attribute and computes information gain. The attribute having more information gain with less entropy is considered the root node that is responsible for splitting.

### <a id="gini-index" name="gini-index">9.2.2 Gini Index</a>

<a href = "#top">⇑ table of contents</a>

Another attribute selection measure is *gini index*. This is used by *CART* which is used to create split points. Gini index is given by following formula;

image.png

The Gini index is used to measure the homogeneity or impurity of the dataset with target values (success and failure). The Gini index of the dataset should be less which indicates less impurity and more homogeneity.

The sum of the squares of the probabilities of each class in a node **`(1 - p^2 - q^2)`**, where p and q represent the probabilities of success and failure, yields the node's Gini index.
Based on the percentage of data points in each node, the weighted Gini score of each split node is computed.

The algorithm selects the attribute with the lowest Gini index for the split when dealing with discrete-valued attributes by evaluating subsets according to the Gini index.
The algorithm looks at pairs of adjacent values as possible split points for continuous-valued attributes. The split that has the lowest Gini index is selected.

So, in the end, the attribute with a minimum Gini index or high information gain is chosen as the decision node.

### <a id="overfitting" name="overfitting">9.2.3 Overfitting in decision tree</a>

<a href = "#top">⇑ table of contents</a>

Overfitting is the most common problem encountered while dealing with decision trees. It arises due to the evolution of many branches of the tree due to outliers and irregularities in data. We can handle overfitting with the help of the following methodologies;

- pre-pruning
- post-pruning

**Pre-pruning** involves stopping the tree-building process a little early. If a node's goodness measure is below a certain threshold, we would rather not split it. But figuring out where to stop is a tough decision.

In order to create a full tree, we prune further and further into the tree. Pruning is done as a **post-pruning** procedure if the tree exhibits the overfitting issue. We assess the effectiveness of our pruning using the cross-validation data. We investigate whether or not expanding a node will lead to improvement using **cross-validation data**. We can move on and expand that node if it indicates an improvement. Yet, it shouldn't be expanded if it indicates a decline in accuracy. Thus, the node ought to be changed into a leaf node.

```python
import category_encoders as ce

encoder = ce.OrdinalEncoder(cols=['buying','maintaince','doors','persons','lug_boot','safety'])
X_train = encoder.fit_transform(X_train)
X_test = encoder.fit_transform(X_test)
```

## <a id="DT" name="DT">9.5 Decision Tree algorithms</a>

We have discussed the "Gini index" and "entropy" in our last sections. Now we'll use these criteria to create a machine-learning model.

### <a id="dt-gini" name="dt-gini">9.5.1 Decision tree with gini index</a>

<a href = "#top">⇑ table of contents</a>

```python
from sklearn.tree import DecisionTreeClassifier

tree_gini = DecisionTreeClassifier(criterion='gini',max_depth=3,random_state=0)
tree_gini.fit(X_train,y_train)
```

```python
tree_pred_gini = tree_gini.predict(X_test)
```

We have imported **`DecisionTreeClassifier`** because the problem we are dealing with is a classification kind of problem.

Let's see the accuracy of the model.

```python
from sklearn.metrics import accuracy_score

tree_score_gini = accuracy_score(y_test,tree_pred_gini)
print(f'accuracy score over test set: {tree_score_gini}')
```

The accuracy of the model over the test set is 72.63%. Now let's check the accuracy score over the train set.

```python
tree_pred_train = tree_gini.predict(X_train)
```

```python
tree_score_train = accuracy_score(y_train,tree_pred_train)
print(f'accuracy score over train set: {tree_score_train}')
```

As the accuracy of the model over the training and the test is comparable there is no sign of overfitting.

Let's see wha our decision tree looks like. The below code helps you to plot the decision tree visually.

```python
from sklearn import tree

tree.plot_tree(tree_gini.fit(X_train,y_train))
```

This is what the decision tree looks like for the machine learning model. For every node, it has calculated the Gini index, and the attribute with the least Gini index is considered the root node of the tree. The tree has only 3 levels.

Let's visualize the decision tree using another method called "*graphviz*".

```python
import graphviz

dot_data = tree.export_graphviz(tree_gini,
                             out_file =None,
                             feature_names=X_train.columns,
                             class_names = y_train,
                             filled = True,
                             rounded = True,
                             special_characters = True
                             )

graph = graphviz.Source(dot_data)
graph
```

Let's do the same procedure but this time using "entropy" criteria.
