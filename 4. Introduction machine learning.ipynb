## <a id="machine-learning" name="machine-learning" >4.1 What is Machine Learning?</a>

The science and art of teaching computers to learn from data are combined in the multidisciplinary is called the field of **machine learning**. This involves developing models and algorithms that allow systems to automatically get better over time at a particular task.

The ability of computers to learn without explicit programming is known as **machine learning**, and it emphasizes the transition from conventional rule-based systems to more intelligent and adaptable ones.

We have one more definition which is very formal,

A computer program is said to learn from experience **E** with respect to some task **T** and some performance measure P if its performance on T, as measured by **P**, improves with experience **E**.


Let's discuss with an example.

image.png

A best application of machine learning is speech recognition. In this case, the experience E is the set of audio samples and their corresponding transcriptions (training set), the task T is to transcribe spoken words into text, and the performance measure P could be the accuracy of accurately transcribing new spoken phrases.

 Here, the underlying algorithm or architecture that learns and makes predictions is referred to as the **model**. Speech recognition models can be created using neural networks, such as transformer models or recurrent neural networks (RNNs).

 The system is trained by exposing it to different speech patterns, which helps the system learn and become more accurate at transcribing over time.


Let's look at some of the important terminology we should know throughout this machine learning.

* **Model**
> A machine learning system's model is its fundamental building block for extracting patterns from data. It displays the relationships that have been discovered between the target variable and the input features.


* **Training set**
 > A portion of the data used to train a machine learning model is called the training set. Input-output pairs are utilized to instruct the model on patterns and relationships.


* **Test set**
>A distinct subset of the data that was not seen by the model during training is known as the test set. It is employed to appraise the model's generalization capacity and performance on untested data.


* **Features**
> A feature is a specific, measured attribute or feature of the data. The machine learning model uses features as input variables.


* **Label or target variable**
> The variable that the model seeks to predict is known as the label or target variable. The model is trained to map input features to the target variable in supervised learning.


* **Hyperparameter**
> Hyperparameters are external model configuration parameters that are not determined by the data. Learning rates, regularization strengths, and the quantity of hidden layers in a neural network are a few examples.


* **overfitting**
> When a model learns the training set too well—capturing noise or erratic fluctuations—overfitting takes place. Poor generalization on fresh, untested data may result from this.


* **Underfitting**
> Underfitting happens when a model is too simple to capture the underlying patterns in the data, resulting in poor performance on both the training and test sets.

Just try to remember these terminologies. Whenever we learn something we have to go with the terminologies that are introduced.

There are so many types of machine learning systems that are divided based on few criteria.

### <a id="supervised" name="supervised">4.3.1 Supervised Learning</a>
> Supervised learning is a strategy where we provide the *labels* in the data set along with the *features* to the algorithm. There are generally two kinds of problems we deal with when we are working on the supervised learning algorithm. They are;
- Classification
- Regression

image.png

**Regression** predicts numerical values; **classification** places data points into specified groups. These are two kinds of supervised machine learning, which use algorithms to examine and identify patterns in labeled training data. This allows for precise predictions to be made in a variety of industries, including marketing, finance, and healthcare.

The best example for regression is 'House price prediction' where we need to output the price of the house based on the input features we give and for classification 'Spam filter' bis the est example.


### <a id="unsupervised" name="unsupervised">4.3.2 Unsupervised Learning</a>
>  In unsupervised learning, we don't provide any kind of labels to the data. This learning is just like knowing or learning something without any teacher.

Without preset results, unsupervised learning investigates patterns in unlabeled data. Algorithms that identify innate structures, relationships, or similarities include clustering and association. Unsupervised learning, which is widely applied in anomaly detection, pattern recognition, and data exploration, improves insights by exposing hidden structures in a variety of datasets.


image.png

### <a id="semi" name="semi">4.3.3 Semi-supervised Learning</a>
> Labeled and unlabeled data are combined in semi-supervised learning to train models. It provides a useful method for training models with little labeled data by combining the two forms of information, which strikes a compromise between the effectiveness of supervised learning and the scalability of unsupervised learning.

### <a id="reinforcement" name="reinforcement">4.3.4 Reinforcement Learning</a>
> In the machine learning paradigm known as reinforcement learning, agents pick up knowledge by interacting with their surroundings. By making mistakes and learning from them, the agent can improve its decision-making to accomplish objectives by receiving feedback in the form of incentives or penalties. Widely used in autonomous systems, robotics, and gaming, it permits intelligent and adaptable behavior.

## <a id="challenges" name="challenges">4.4 Challenges in machine learning</a>

Being a machine learning engineer, our task is to clean the data, get insights and get a ML model prepared. When we do so, we can fall into these two faces:
- bad data
- bad model

Let's look at some of the challenges of the machine learning;
- *insufficient quantity of data*
> One of the biggest problems in machine learning is the lack of training data. Insufficient data may cause models to overfit to the available samples, making it difficult for them to generalize to new cases.

- *non-representative training data*
> Machine learning is made more difficult by non-representative training data since models may learn skewed patterns that don't accurately reflect the diversity of the target population.

- *poor quality in data*
> Machine learning faces a significant challenge from low-quality data. Unreliable predictions can be produced by models that are misled by noisy, incomplete, or inaccurate data.

- *irrelevant features*
> The data that we are providing should have proper features. There should be no irrelevant features that could be nonsense to the ML model.  Primarily, we need to focus on;
>>- feature selection
>>- feature extraction
>>- creating new features

- *Overfitting*
> A common problem in machine learning is called overfitting, which occurs when a model learns the training data too well and captures noise or random fluctuations instead of the underlying patterns. Consequently, the model exhibits remarkable performance on the training set but is unable to generalize to new, unobserved data.

- *underfitting*
> In machine learning, underfitting happens when a model is too basic to identify the underlying patterns in the data. Because it lacks the complexity required to accurately represent the relationships within the dataset, it performs poorly on both training and new data.
